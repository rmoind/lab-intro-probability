#### Ironhack Airlines 

Often Airlines sell more tickets than they have seats available, this is called overbooking. Consider the following:
- A plane has 450 seats. 
- Based on historical data we conclude that each individual passenger has a 3% chance of missing it's flight. 

If the Ironhack Airlines routinely sells 460 tickets, what is the chance that they have a seats for all passenger?

```python
# Plane seats: 450
# Tickets sold: 460
# Probability a passenger misses flight: 3% 
# p = 0.03


# This is a binomial probability problem. 
# the probability that at least 10 passengers miss the flight (since 460 - 10 = 450).
#So, P(X ≥ 10) = 1 - P(X ≤ 9), where X ~ Binomial(n=460, p=0.03).

from scipy.stats import binom

n = 460
p = 0.03
prob = 1 - binom.cdf(9, n, p)
1 - binom.cdf(9, n, p)
```

```python
#The probability that all passengers who show up can be seated is approximately 0.8845 or 88.45%.
```

What is the probability that the representative needs to make at least three attempts before successfully resolving a customer complaint?

```python
#Probability of resolving a complaint on first attempt: p = 0.3
#This is a geometric distribution problem.
#P(X ≥ 3) = (1 - p)^(3-1) = (0.7)^2

from scipy.stats import geom

p = 0.3
prob = geom.sf(2, p)  # P(X ≥ 3) = P(X > 2)
geom.sf(2, p)

```

```python
#The probability that the representative needs at least three attempts before resolving a complaint is approximately 0.49 or 49%.
```

**Challenge 3**

#### Ironhack Website

Consider a scenario related to Ironhack website traffic. Where:
- our website takes on average 500 visits per hour.
- the website's server is designed to handle up to 550 vists per hour.


What is the probability of the website server being overwhelmed?

```python
#Average visits per hour: 500
#Server capacity: 550 visits/hour

#This is a Poisson distribution problem.
#P(X > 550) for X ~ Poisson(λ=500)
#P(X > 550) for at least one hour in 24: 1 - P(X ≤ 550 for all 24 hours)

from scipy.stats import poisson

# 1. in one hour
lambda_ = 500
prob_hour = 1 - poisson.cdf(550, lambda_)
prob_hour
```

```python
#The probability that the server is overwhelmed in one hour is approximately 0.01289 or 1.29%.
```

What is the probability of being overwhelmed at some point during a day? (consider 24hours)

```python
# What is the probability of being overwhelmed at some point during a day (24 hours)?

from scipy.stats import poisson

# 2. Probability in 24 hours
prob_not_overwhelmed_hour = poisson.cdf(550, lambda_)
prob_not_overwhelmed_day = prob_not_overwhelmed_hour**24
prob_overwhelmed_day = 1 - prob_not_overwhelmed_day
prob_overwhelmed_day
```

```python
#The probability that the server is overwhelmed in one hour is approximately 0.2677 or 26.77%.
```

#### Ironhack Helpdesk

Consider a scenario related to the time between arrivals of customers at a service desk.

On average, a customers arrives every 10minutes.

What is the probability that the next customer will arrive within the next 5 minutes?

```python
#Average time between arrivals: 10 minutes (λ = 1/10 per minute)
#Probability next customer arrives within 5 minutes.

#This is an exponential distribution problem.
#P(X ≤ 5) for X ~ Exp(λ=1/10)

from scipy.stats import expon

lambda_ = 1/10  # Rate parameter (λ)
scale = 1/lambda_  # Scale parameter (1/λ)

expon_dist = expon(scale=scale)
prob_within_5 = expon_dist.cdf(5)
prob_within_5
```

```python
#The probability that the next customer arrives within 5 minutes is approximately 0.3935 or 39.35%.
```

If there is no customer for 15minutes, employees can that a 5minutes break.

What is the probability an employee taking a break?

```python
#Probability of no customer for 15 minutes (so employees can take a break).

#Approach:
#P(X > 15) for X ~ Exp(λ=1/10)

prob_no_customer_15 = 1 - expon.cdf(15, scale=scale)
prob_no_customer_15
```

```python
#The probability of no customer for 15 minutes is approximately 0.2231 or 22.31%.
```

**Challenge 5**

The weights of a certain species of birds follow a normal distribution with a mean weight of 150 grams and a standard deviation of 10 grams. 

- If we randomly select a bird, what is the probability that its weight is between 140 and 160 grams?

```python
#Mean weight: 150g
#Standard deviation: 10g

#Probability a randomly selected bird weighs between 140g and 160g.

#This is a normal distribution problem.
#P(140 ≤ X ≤ 160) for X ~ N(150, 10^2)
from scipy.stats import norm

mu, sigma = 150, 10
prob = norm.cdf(160, mu, sigma) - norm.cdf(140, mu, sigma)
prob
```

```python
#The probability that a randomly selected bird weighs between 140g and 160g is approximately 0.6826 or 68.26%.
```

If the lifetime (in hours) of a certain electronic component follows an exponential distribution with a mean lifetime of 50 hours, what is the probability that the component fails within the first 30 hours?

```python
# Mean lifetime: 50 hours (λ = 1/50 per hour)

#Probability the component fails within the first 30 hours.
#Approach:
# P(X ≤ 30) for X ~ Exp(λ=1/50)

from scipy.stats import expon

lambda_ = 1/50
scale=1/lambda_

prob = expon.cdf(30, scale=scale)
prob
```

```python
#The probability that the component fails within the first 30 hours is approximately 0.4512 or 45.12%.
```
